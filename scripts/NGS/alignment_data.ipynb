### Inspect the headers

```python
file = "/Users/emrecanciftci/betik/2024_bio/large_data/BAM_files/NA18489.chrom20.ILLUMINA.bwa.YRI.exome.20121211.bam"
bam = pysam.AlignmentFile(file, mode= "rb")

headers = bam.header

for record_type, records in headers.items():
    print(record_type)
    for i, record in enumerate(records):
        if type(record) == dict:
            print('\t%d' % (i + 1))
            for field, value in record.items():
                print('\t\t%s\t%s' % (field, value))

        else:
            print('\t\t%s' % record)
```
